# Environmental WENR features

Within the WEED project we have made the WENR features available for openEO as an alpha release. These WENR features contain information on the european soil, population and climate over a timespan of  1981-2018. Caution, the exact temporal period depends from band to band. More information on this collection can be found here: https://radiantearth.github.io/stac-browser/#/external/catalogue.weed.apex.esa.int/collections/wenr_features?.language=en 

```python
import openeo

connection = openeo.connect("openeo.dataspace.copernicus.eu").authenticate_oidc()
datacube = connection.load_stac("https://catalogue.weed.apex.esa.int/collections/wenr_features",
                                spatial_extent={"west": 5.0, "south": 51.0, "east": 5.5, "north": 51.5, "epsg": 4326})

datacube.execute_batch('test.nc')
```

Once the job has been completed, we can visualise the various bands:

```python
import xarray as xr
import matplotlib.pyplot as plt

ds = xr.open_dataset('test.nc')

plt.figure()
ds['soc'].isel(t=0).plot()

plt.title('Soil Organic Carbon Topsoil')
plt.tight_layout()
plt.show()
```
